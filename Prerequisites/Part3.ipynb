```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
import time
import matplotlib.pyplot as plt
```

```python
train_set = pd.read_csv('train.csv')
test_set = pd.read_csv('test.csv')

x_train = (train_set.drop(['median_house_value'], axis=1))
y_train = np.array(train_set['median_house_value'])

x_test = np.array(test_set.drop(['median_house_value'], axis=1))
y_test = np.array(test_set['median_house_value'])
```

```python
x_train.info()
```

```python
model = LinearRegression()
start = time.time()
model.fit(x_train,y_train)
end = time.time()
print("Training time: ", end - start)
y_hat = model.predict(x_test)
```

```python
print("Mean absolute error:", mean_absolute_error(y_test, y_hat))
print("Mean squared error:", mean_squared_error(y_test, y_hat))
print("R² score:", r2_score(y_test, y_hat))
```

```python
plt.scatter(y_test, y_hat, alpha=0.07)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r', lw=1)
plt.xlabel("Actual Median House Value")
plt.ylabel("Predicted Median House Value")
plt.title("Actual vs Predicted House Values")
plt.show()
```

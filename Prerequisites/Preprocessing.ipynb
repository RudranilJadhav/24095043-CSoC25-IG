```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('housing.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df=df.join(pd.get_dummies(df.ocean_proximity))
df = df.replace({True: 1, False: 0})
X=df.drop(['median_house_value','ocean_proximity'], axis=1)
Y=df['median_house_value']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25)
```

```python
min_vals = x_train.min()
max_vals = x_train.max()

x_train = (x_train - min_vals) / (max_vals - min_vals)
x_test = (x_test - min_vals) / (max_vals - min_vals)
```

```python
train_set= x_train.copy()
train_set['median_house_value']=y_train

test_set = x_test.copy()
test_set['median_house_value']=y_test

train_set
```

```python
train_set.corr()['median_house_value']
```

```python
epsilon = 1e-5

train_set['total_bedrooms'] = train_set['total_bedrooms'].replace(0, epsilon).fillna(epsilon)
train_set['population'] = train_set['population'].replace(0, epsilon).fillna(epsilon)

train_set['rooms/bedrooms']=train_set['total_rooms']/train_set['total_bedrooms']
train_set['household/population']=train_set['households']/train_set['population']
train_set = train_set.drop(['population','households','total_bedrooms'], axis=1)

test_set['rooms/bedrooms']=test_set['total_rooms']/test_set['total_bedrooms']
test_set['household/population']=test_set['households']/test_set['population']
test_set = test_set.drop(['population','households','total_bedrooms'], axis=1)

train_set.corr()['median_house_value']
```
